```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

## FUNCTIONAL

```python


plt.plot(x,y,);
plt.xlabel('X Label')
plt.ylabel('y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b');
```

## OBJECT ORIENTED METHOD

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8,0.8])

axes.plot(x,y);
axes.set_xlabel('X Label')
axes.set_ylabel('y Label')
axes.set_title('Title');
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y, 'g')
axes2.plot(y,x, 'r');

axes1.set_title('LARGER TITLE');
axes2.set_title('SMALLER TITLE')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y, 'r');
axes[0].set_title('FIRST PLOT')

axes[1].plot(y,x, 'g')
axes[1].set_title('SECOND PLOT')

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
```

## FIGSIZE AND DPI

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'r');
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10,6))

axes[0].plot(x,y, 'b');
axes[1].plot(y,x, 'r');

plt.tight_layout()
```

```python
fig.savefig('my_picture.jpeg', dpi=400)
```

```python
fig = plt.figure(figsize=(9,6))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_ylabel('y Label')
ax.set_xlabel('X Label');
```

## LEGEND

```python
fig = plt.figure(figsize=(9,6))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed');
ax.legend(loc='best');
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=1, alpha=1, ls='-', marker='o', ms=20, 
        markerfacecolor='yellow', markeredgecolor='black', markeredgewidth=3); 
# RGB Hex Code, # lw: linewidth # ls: linestyle #ms:markersize
```
